```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('brain_pic.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
blur = cv2.GaussianBlur(img,(3,3),100)
```

```python
plt.imshow(blur)
```

```python
thr = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,205, 20)
plt.imshow(thr, cmap='gray')
```
